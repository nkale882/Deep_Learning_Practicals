```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
 # Load the image
img_path = '/content/Cartoons.jpg'
img = cv2.imread(img_path)
```

```python
print(img)
print("Image shape:", img.shape)
```

```python
# Show the original image
plt.imshow(img)
plt.title("Original Image")
plt.axis("on")
plt.show()
```

```python
#  Cropping the image
crop = img[0:200, 150:300]
# Display cropped image
plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("on")
plt.show()
```

```python
img[0:,0:100]=[200,90,0]
img[0:,350:]=[200,90,0]
plt.imshow(img)
```
